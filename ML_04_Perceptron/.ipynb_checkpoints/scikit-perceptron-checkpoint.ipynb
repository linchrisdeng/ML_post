```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bokeh.io import output_notebook, output_file
from bokeh.plotting import figure, output_file, show, reset_output
from bokeh.models import HoverTool, Label, WheelZoomTool, PanTool
from bokeh.layouts import row
```

```python
df = pd.read_csv("df.csv", header=None)
df.columns = ['y', 'x']
df["lable"] = np.where(df.x > 0, 1, -1)
df_1 = df.loc[df.x > 0]
df_2 = df.loc[df.x < 0]
```

```python
reset_output()
output_notebook()
TOOLTIPS = [
    ("x", "@x"),
    ("y", "@y")]

p = figure(plot_height = 300, plot_width = 600,
            tools = "crosshair")
p.add_tools(HoverTool(tooltips = TOOLTIPS, mode = 'vline'))
p.circle(x = df_1.x, y = df_1.y,color ='red')
p.circle(x = df_2.x, y = df_2.y, color = 'blue')
p.toolbar.logo = None
show(p)
```

```python
x = np.array(df.iloc[:,0:2])
x.shape = (len(x), 2)
ones = np.ones((len(x), 1))
X = np.concatenate((ones, x), axis = 1)
y = df.lable
```

```python
from mpl_toolkits import mplot3d
%matplotlib widget

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i, m in [(df_1, 'o'), (df_2, '^')]:
    zs = i.lable
    xs = i.x
    ys = i.y
    ax.scatter(xs, ys, zs, marker = m)

plt.show()
```

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import confusion_matrix
clf = Perceptron(tol=1e-3, random_state = 1)
clf.fit(x, y)
clf.score(x,y)

xx = np.linspace(np.min(x[:,1]), np.max(x[:,1]),100)
yy = np.linspace(np.max(x[:,0]), np.min(x[:,0]),100)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack(np.c_[XX.ravel(), YY.ravel()])
Z = clf.predict(xy).reshape(XX.shape)

trace2 = go.Surface(
    x=XX,
    y=YY,
    z=Z,
    opacity=0.4,
    showscale = False
)
```
