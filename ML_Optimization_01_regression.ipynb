```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy.linalg import inv

from bokeh.io import output_notebook
from bokeh.plotting import figure, output_file, show
from bokeh.models import HoverTool, Label, WheelZoomTool, PanTool
from bokeh.layouts import row
output_notebook()
```

Load up the dataset.

```python
# import the dataset
data = np.asarray(pd.read_csv('student_debt_data.csv',header = None))
x = data[:,0]
x.shape = (len(x),1)
y = data[:,1]
y.shape = (len(y),1)

# pad input with ones
o = np.ones((len(x),1))
x_new = np.concatenate((o,x),axis = 1)


w =np.matmul(inv(np.matmul(x_new.T, x_new)), np.matmul(x_new.T, y))      # weights learned by solving linear system
w
```

```python
# plot data with linear fit - this is optional
s = np.linspace(np.min(x),np.max(x), 40)
t = w[0] + w[1]*s
p1 = figure(plot_height = 300, plot_width = 300,
            tools = "crosshair")
p2 = figure(plot_height = 300, plot_width = 300,
            tools = "crosshair")
p1.circle(x = x.flatten(), y = y.flatten())

p2.line(x = s, y = t, color = "red")
p2.circle(x = x.flatten(), y = y.flatten())
p1.toolbar.logo = None
p2.toolbar.logo = None

show(row(p1, p2))
output_file("Reg_00.html")
```
